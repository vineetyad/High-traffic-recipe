```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats import boxcox, yeojohnson
from scipy.stats.mstats import winsorize
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Read recipe site traffic data file and convert it into a dataframe
recipe_traffic_data = pd.read_csv("recipe_site_traffic_2212.csv")

# Checking basic information about the dataframe
recipe_traffic_data.head()
recipe_traffic_data.info()

# Data preprocessing: handling servings column
recipe_traffic_data['servings'] = recipe_traffic_data['servings'].str.replace(" as a snack", "")
recipe_traffic_data['servings'] = recipe_traffic_data['servings'].astype('int')

# Data preprocessing: handling high traffic column
recipe_traffic_data['high_traffic'] = np.where(recipe_traffic_data['high_traffic'] == "High", True , False)

# Data preprocessing: handling category column
recipe_traffic_data['category'] = recipe_traffic_data['category'].str.replace(" Breast", "")
recipe_traffic_data['category'] = recipe_traffic_data['category'].astype('category')

# Checking for missing values and handling them
recipe_traffic_data = recipe_traffic_data.dropna().reset_index(drop=True)
recipe_traffic_data.info()

# Data visualization: 
numeric_vars = ['calories', 'carbohydrate', 'sugar', 'protein']
categorical_vars = ['category', 'servings', 'high_traffic']

fig, ax = plt.subplots(3, 3, figsize=(15, 20))

# Plotting the numeric variables
for i, var in enumerate(numeric_vars):
    sns.histplot(recipe_traffic_data[var], kde=False, bins=30, ax=ax[i//3, i%3])
    ax[i//3, i%3].set_title(f'Distribution of {var}')

# Plotting the categorical variables
for i, var in enumerate(categorical_vars, start=4):
    sns.countplot(x=var, data=recipe_traffic_data, ax=ax[i//3, i%3])
    ax[i//3, i%3].set_title(f'Count of {var}')
    if var == 'category':
        for label in ax[i//3, i%3].get_xticklabels():
            label.set_rotation(90)

# Remove empty subplots
for i in range(len(numeric_vars) + len(categorical_vars), 9):
    ax[i//3, i%3].axis('off')

plt.tight_layout()
plt.show()

# Add high_traffic to the list of numeric variables
numeric_vars = ['calories', 'carbohydrate', 'sugar', 'protein', 'high_traffic']

# Correlation heatmap
correlation_matrix = recipe_traffic_data[numeric_vars].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True)
plt.title('The Correlation Heatmap between Numeric Variables')
plt.show()

# Boxplots
numeric_vars = ['calories', 'carbohydrate', 'sugar', 'protein']

fig, ax = plt.subplots(2, 3, figsize=(15, 10)) # 2 rows, 3 columns
for i, var in enumerate(numeric_vars):
    row = i // 3  # integer division gives us the row index
    col = i % 3   # modulus operator gives us the column index
    sns.boxplot(x=recipe_traffic_data['high_traffic'], y=recipe_traffic_data[var], ax=ax[row, col],palette="Set3")
    ax[row, col].set_title(f'high_traffic vs {var}')
fig.delaxes(ax[1, 2]) # this will remove empty subplots
fig.delaxes(ax[1, 1])

plt.tight_layout()
plt.show()

# Data visualization: Count of high traffic recipes by servings
plt.figure(figsize=(12, 6))
sns.countplot(y="servings", hue="high_traffic", data=recipe_traffic_data)
plt.xlabel("Count")
plt.ylabel("Servings")
plt.title("Count of High Traffic Recipes by Servings")
plt.legend(title="High Traffic")
plt.xticks(rotation=90)
plt.show()

# Data visualization: Count of high traffic recipes by category
plt.figure(figsize=(12, 6))
sns.countplot(y="category", hue="high_traffic", data=recipe_traffic_data)
plt.xlabel("Count")
plt.ylabel("Category")
plt.title("Count of High Traffic Recipes by Category")
plt.legend(title="High Traffic")
plt.xticks(rotation=90)
plt.show()

# Data visualization: Normalized count of high traffic recipes by servings
high_traffic_servings = pd.pivot_table(recipe_traffic_data, index=["servings"], columns=["high_traffic"], values="recipe", aggfunc='count')
high_traffic_servings_normalized = high_traffic_servings.div(high_traffic_servings.sum(axis=1), axis=0)
high_traffic_servings_normalized.plot(kind='bar', stacked=True)
plt.xlabel("Servings")
plt.ylabel("Percentage")
plt.title("Normalized Count of High Traffic Recipes by Servings")
plt.show()

# Data visualization: Normalized count of high traffic recipes by category
high_traffic_category = recipe_traffic_data.pivot_table(index="category", columns="high_traffic", values="servings", aggfunc="sum")
normalized_table = high_traffic_category.div(high_traffic_category.sum(axis=1), axis=0)
normalized_table.plot(kind='bar', stacked=True)
plt.xlabel("Category")
plt.ylabel("Percentage")
plt.title("Normalized Count of High Traffic Recipes by Category")
plt.show()

# Function to calculate outlier limits
def calculate_outlier_limits(data, factor=1.5):
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    lower_limit = q1 - (factor * iqr)
    upper_limit = q3 + (factor * iqr)
    return lower_limit, upper_limit

# Calculate outlier limits for numerical columns
numerical_columns = ['calories', 'carbohydrate', 'sugar', 'protein']
outlier_limits = {}
for column in numerical_columns:
    lower_limit, upper_limit = calculate_outlier_limits(recipe_traffic_data[column])
    outlier_limits[column] = (lower_limit, upper_limit)

# Filtering outliers using Yeo-Johnson transformation
transformed_data = recipe_traffic_data.copy()
for column, limits in outlier_limits.items():
    transformed_data[column] = yeojohnson(transformed_data[column])[0]

# Visualizing distributions after transformation
plt.figure(figsize=(16, 4))
for i, column in enumerate(numerical_columns):
    plt.subplot(1, len(numerical_columns), i+1)
    sns.histplot(data=transformed_data, x=column, kde=True)
    plt.title(f"Distribution of {column.capitalize()}")
plt.tight_layout()
plt.show()

# One-hot encoding for the category column
encoded_categories = pd.get_dummies(recipe_traffic_data["category"])
recipe_traffic_data_encoded = pd.concat([transformed_data, encoded_categories], axis=1)
recipe_traffic_data_encoded = recipe_traffic_data_encoded.drop("category", axis=1)

# Splitting data into features and target variable
X = recipe_traffic_data_encoded.drop("high_traffic", axis=1)
y = recipe_traffic_data_encoded["high_traffic"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)

# Function to develop and evaluate models
def develop_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    model_pred_train = model.predict(X_train)
    model_pred_test = model.predict(X_test)
    model_accuracy_train = accuracy_score(y_train, model_pred_train)
    model_precision_train = precision_score(y_train, model_pred_train)
    model_recall_train = recall_score(y_train, model_pred_train)
    model_f1_train = f1_score(y_train, model_pred_train)
    model_cm_train = confusion_matrix(y_train, model_pred_train)
    model_accuracy_test = accuracy_score(y_test, model_pred_test)
    model_precision_test = precision_score(y_test, model_pred_test)
    model_recall_test = recall_score(y_test, model_pred_test)
    model_f1_test = f1_score(y_test, model_pred_test)
    model_cm_test = confusion_matrix(y_test, model_pred_test)
    return {
        'train': [model_accuracy_train, model_precision_train, model_recall_train, model_f1_train, model_cm_train],
        'test': [model_accuracy_test, model_precision_test, model_recall_test, model_f1_test, model_cm_test]
    }

# Model development and evaluation
logreg = LogisticRegression()
logreg_evaluation_metrics = develop_model(logreg, X_train, y_train, X_test, y_test)

dtree = DecisionTreeClassifier()
dtree_evaluation_metrics = develop_model(dtree, X_train, y_train, X_test, y_test)

rf = RandomForestClassifier()
rf_evaluation_metrics= develop_model(rf, X_train, y_train, X_test, y_test)

svm = SVC()
svm_evaluation_metrics = develop_model(svm, X_train, y_train, X_test, y_test)

# Function to print evaluation metrics
def print_metrics(evaluation_metrics):
    print("  Accuracy:", evaluation_metrics[0])
    print("  Precision:", evaluation_metrics[1])
    print("  Recall:", evaluation_metrics[2])
    print("  F1-Score:", evaluation_metrics[3])
    print("  Confusion Matrix:")
    print(evaluation_metrics[4])

# Printing evaluation metrics for Logistic Regression
print("Logistic Regression Train:")
print_metrics(logreg_evaluation_metrics['train'])
print("Logistic Regression Test:")
print_metrics(logreg_evaluation_metrics['test'])

# Printing evaluation metrics for Decision Tree
print("Decision Tree Train:")
print_metrics(dtree_evaluation_metrics['train'])
print("Decision Tree Test:")
print_metrics(dtree_evaluation_metrics['test'])

# Printing evaluation metrics for Random Forest
print("Random Forest Train:")
print_metrics(rf_evaluation_metrics['train'])
print("Random Forest Test:")
print_metrics(rf_evaluation_metrics['test'])

# Printing evaluation metrics for Support Vector Machines
print("Support Vector Machines Train:")
print_metrics(svm_evaluation_metrics['train'])
print("Support Vector Machines Test:")
print_metrics(svm_evaluation_metrics['test'])

# Function to calculate High Traffic Conversion Rate
def calculate_kpi(model_metrics, model_name):
    kpi_train = model_metrics['train'][4]
    kpi_test = model_metrics['test'][4]
    print(f"High Traffic Conversion Rate for {model_name} train: ", kpi_train[1][1]/(kpi_train[0][1]))
    print(f"High Traffic Conversion Rate for {model_name} test: ", kpi_test[1][1]/(kpi_test[0][1]))

# Calculating High Traffic Conversion Rate for each model
calculate_kpi(logreg_evaluation_metrics, "Logistic Regression")
calculate_kpi(dtree_evaluation_metrics, "Decision Tree")
calculate_kpi(rf_evaluation_metrics, "Random Forest")
calculate_kpi(svm_evaluation_metrics, "Support Vector Machines")

```
